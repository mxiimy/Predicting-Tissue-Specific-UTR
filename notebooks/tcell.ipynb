```python
import pandas as pd
from pathlib import Path

tcell_path = Path('../data/tcell_specific_genes.csv')
initialDB_path = Path('../data/initialDB_sequences.csv')

df = pd.read_csv(tcell_path)
initialDB_df = pd.read_csv(initialDB_path)
```

```python
initialDB_lookup = initialDB_df.assign(gene_id=initialDB_df['gene_id'].str.split('.').str[0]).set_index('gene_id')

df["5' UTR"] = df['ensembl_id'].map(initialDB_lookup["5' UTR"])
df['CDS'] = df['ensembl_id'].map(initialDB_lookup['CDS'])
df["3' UTR"] = df['ensembl_id'].map(initialDB_lookup["3' UTR"])

df 
```
