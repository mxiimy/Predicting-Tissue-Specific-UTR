```python
import os
import pandas as pd
import numpy as np
import re
from pathlib import Path

initialDB_path = Path('../data/initialDB_sequences.csv')
initialDB = pd.read_csv(initialDB_path, sep=',', header=0, comment='#', names=[
    'gene_id', 'transcript_id', "5' UTR", 'start_codon', 'CDS', 'stop_codon', "3' UTR", 'appris_rank', "5' UTR_len", 'CDS_len', "3' UTR_len"
])

df = initialDB.copy()
df.head()
```

```python
df.drop(columns=['gene_id', "5' UTR_len", 'CDS_len', "3' UTR_len"], inplace=True)

df.head()
```

```python
df = initialDB.copy()
df['sequence'] = (
    df["5' UTR"].fillna('').astype(str)
    + df['CDS'].fillna('').astype(str)
    + df["3' UTR"].fillna('').astype(str)
)
df.drop(columns=['gene_id', '5\' UTR', 'CDS', '3\' UTR', 'start_codon', 'stop_codon', 'appris_rank', '5\' UTR_len', 'CDS_len', '3\' UTR_len'], inplace=True)
df.rename(columns={'transcript_id': 'id'}, inplace=True)

df.head()
```

```python
df.to_csv('../data/ARMnet_sequences.csv', index=False)
```
