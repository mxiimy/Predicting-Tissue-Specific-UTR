```python
import os
import glob

RBP_DIR = "/lustre09/project/6007512/HeDS/melody/Predicting-Tissue-Specific-UTR/"
MASTER_CSV_PATH = "/lustre09/project/6007512/HeDS/melody/Predicting-Tissue-Specific-UTR/data/finalDB_with_tai_local.csv"
OUTPUT_FILE = "/lustre09/project/6007512/HeDS/melody/Predicting-Tissue-Specific-UTR/data/finalDB_with_all_rbps.csv"

# CisBP metadata file path 
RBP_INFO_METADATA = os.path.join(RBP_DIR, "data/RBP_Information.txt")

# Generate the Dictionary Mapping from CisBP Metadata
rbp_name_map = {}

if os.path.exists(RBP_INFO_METADATA):
    print(f"Parsing metadata map from {RBP_INFO_METADATA}...")
    try:
        # CisBP files are typically tab-separated text files
        meta_df = pd.read_csv(RBP_INFO_METADATA, sep='\t')
        
        # CisBP columns are standardly labeled 'Motif_ID' and 'RBP_Name'
        if 'Motif_ID' in meta_df.columns and 'RBP_Name' in meta_df.columns:
            # Drop rows missing names and build the dictionary pairs
            meta_df = meta_df.dropna(subset=['Motif_ID', 'RBP_Name'])
            rbp_name_map = dict(zip(meta_df['Motif_ID'], meta_df['RBP_Name']))
            print(f"Successfully loaded {len(rbp_name_map)} protein mapping pairs!")
        else:
            print("[WARNING] Could not find 'Motif_ID' or 'RBP_Name' columns. Defaulting to raw file IDs.")
    except Exception as e:
        print(f"[WARNING] Failed to parse metadata file: {e}. Defaulting to raw file IDs.")
else:
    print(f"[INFO] No RBP_Information.txt found at {RBP_INFO_METADATA}. File IDs will remain as M#####_2.00.")


def get_readable_name(filename):
    """Extracts RBP motif code from filename and looks up its readable name."""
    # Handles both 'temp_M00002_2.00.csv' and 'M00002_2.00' formats
    base = os.path.basename(filename).replace("temp_", "").replace(".csv", "").replace(".txt", "")
    return rbp_name_map.get(base, base)


# 3. Build a fast mapping index from the master DB (transcript_id -> gene_id)
print("Loading master mapping database...")
master_base = pd.read_csv(MASTER_CSV_PATH, usecols=['gene_id', 'transcript_id'])
master_base['transcript_id'] = master_base['transcript_id'].astype(str)
master_base['gene_id'] = master_base['gene_id'].astype(str)

# Create an empty matrix shell using only unique gene_ids
final_df = pd.DataFrame({'gene_id': master_base['gene_id'].unique()})

# 4. Find all individual RBP temp files matching your run outputs
rbp_files = glob.glob(os.path.join(RBP_DIR, "temp_*.csv"))
print(f"Found {len(rbp_files)} target RBP chunk files to process.")

# 5. Iteratively process and aggregate each RBP file
for file_path in sorted(rbp_files):
    rbp_name = get_readable_name(file_path)
    print(f"Consolidating scores for: {rbp_name}...")
    
    try:
        # Load single RBP data
        rbp_df = pd.read_csv(file_path)
        rbp_df['transcript_id'] = rbp_df['transcript_id'].astype(str)
        
        # Dynamically discover variable score column names inside the file
        old_3utr_col = [c for c in rbp_df.columns if '3_UTR' in c][0]
        old_5utr_col = [c for c in rbp_df.columns if '5_UTR' in c][0]
        
        # Map transcript_id to gene_id
        rbp_df = rbp_df.merge(master_base, on='transcript_id', how='inner')
        
        # Compress transcripts to Gene Level by pulling the max binding affinity score
        gene_level_rbp = rbp_df.groupby('gene_id').agg({
            old_3utr_col: 'max',
            old_5utr_col: 'max'
        }).reset_index()
        
        # Format headers exactly as requested
        gene_level_rbp = gene_level_rbp.rename(columns={
            old_3utr_col: f"3' UTR score for {rbp_name}",
            old_5utr_col: f"5' UTR score for {rbp_name}"
        })
        
        # Stitch column onto master matrix dataframe
        final_df = final_df.merge(gene_level_rbp, on='gene_id', how='left')
        
    except Exception as e:
        print(f"  Warning: Skipped file {file_path} due to error: {e}")

# 6. Save the matrix
print(f"Saving compiled matrix to {OUTPUT_FILE}...")
final_df.to_csv(OUTPUT_FILE, index=False)
print("All matches parsed and consolidated!")
```

```python
mfe_path = Path('../data/mfe_results.csv')
mfe_db = pd.read_csv(mfe_path, sep=',', header=0, comment='#', names=[
    'gene_name', 'gene_id', 'transcript_id', 'protein_id', "5' UTR", "CDS", "3' UTR", "protein_sequence", "5' UTR_len", "CDS_len", "3' UTR_len", "5' UTR_MFE", "CDS_MFE", "3' UTR_MFE"
])
### Note: MFE values of NaN were not able to be folded by RNAfold either because the sequence is length 0 or it's too long (over 100k nt) for RNAfold to handle.

mfe_db['transcript_id'] = mfe_db['transcript_id'].astype(str)
TAI_DB['transcript_id'] = TAI_DB['transcript_id'].astype(str)

TAI_DB.head()
```

```python
mfe_db = mfe_db.rename(columns={
    '5_UTR_MFE': "5' UTR_MFE",
    '3_UTR_MFE': "3' UTR_MFE"
})

# 2. String strip data keys to ensure safe matches
for dataframe in [mfe_db, TAI_DB]:
    dataframe['transcript_id'] = dataframe['transcript_id'].astype(str).str.strip()
    dataframe['gene_id'] = dataframe['gene_id'].astype(str).str.strip()

# 3. Isoform Fallback: Group by gene_id and fill missing gaps using sibling transcripts
print("Imputing missing transcript gaps from computed gene matches...")
for score_col in ["5' UTR_MFE", "CDS_MFE", "3' UTR_MFE"]:
    # Sort so that non-NaN values shift optimally for grouping containment
    mfe_db = mfe_db.sort_values(by=['gene_id', score_col])
    
    # Forward-fill followed by backward-fill within each gene group
    mfe_db[score_col] = mfe_db.groupby('gene_id')[score_col].ffill().bfill()

# 4. Clean up structural overlaps in destination matrix before stitching
cols_to_clear = [c for c in ["5' UTR_MFE", "CDS_MFE", "3' UTR_MFE"] if c in TAI_DB.columns]
if cols_to_clear:
    TAI_DB = TAI_DB.drop(columns=cols_to_clear)

# 5. Merge seamlessly on transcript_id
TAI_DB = pd.merge(TAI_DB, mfe_db[['transcript_id', "5' UTR_MFE", 'CDS_MFE', "3' UTR_MFE"]], 
                  on='transcript_id', how='left')

print("Verification Layout:")
print(TAI_DB[["gene_id", "transcript_id", "5' UTR_MFE", "CDS_MFE", "3' UTR_MFE"]].head(6))
```

```python
from pathlib import Path
import pandas as pd

mfe_junction_path = Path('../data/mfe_junctions.csv')

# 1. Read naturally using the saved file headers (avoids layout misalignment)
mfe_junction_df = pd.read_csv(mfe_junction_path, sep=',', header=0, comment='#')

# 2. Force matching string data types on index keys for easy downstream merges
mfe_junction_df['transcript_id'] = mfe_junction_df['transcript_id'].astype(str).str.strip()
mfe_junction_df['gene_id'] = mfe_junction_df['gene_id'].astype(str).str.strip()

# Print layout verification to ensure everything lines up beautifully
print(f"[INFO] Loaded database size: {mfe_junction_df.shape}")
print(mfe_junction_df[['transcript_id', '5UTR_CDS_junction_MFE', '5UTR_CDS_junction_length']].head())
```

```python
# 1. Normalize columns in mfe_junction_df to match TAI_DB's nomenclature
mfe_db = mfe_junction_df.rename(columns={
    '5_UTR_MFE': "5' UTR_MFE",
    '3_UTR_MFE': "3' UTR_MFE"
})

# 2. String strip data keys to ensure safe matches
for dataframe in [mfe_db, TAI_DB]:
    dataframe['transcript_id'] = dataframe['transcript_id'].astype(str).str.strip()
    dataframe['gene_id'] = dataframe['gene_id'].astype(str).str.strip()

# 3. Isoform Fallback: Group by gene_id and fill missing gaps using sibling transcripts
print("Imputing missing transcript gaps from computed gene matches...")
target_cols = ["5UTR_CDS_junction_MFE"]

for score_col in target_cols:
    # Ensure column exists in mfe_db before processing fallback
    if score_col in mfe_db.columns:
        # Sort so that non-NaN values shift optimally for grouping containment
        mfe_db = mfe_db.sort_values(by=['gene_id', score_col])
        # Forward-fill followed by backward-fill within each gene group
        mfe_db[score_col] = mfe_db.groupby('gene_id')[score_col].ffill().bfill()

# 4. Clean up structural overlaps in destination matrix before stitching to avoid _x/_y suffixes
cols_to_clear = [c for c in target_cols if c in TAI_DB.columns]
if cols_to_clear:
    TAI_DB = TAI_DB.drop(columns=cols_to_clear)

# 5. Merge seamlessly on transcript_id
# Dynamically select only the columns that actually exist in your processed mfe_db
keep_cols = ['transcript_id'] + [c for c in target_cols if c in mfe_db.columns]
TAI_DB = pd.merge(TAI_DB, mfe_db[keep_cols], on='transcript_id', how='left')

print("Verification Layout:")
available_verify_cols = ['gene_id', 'transcript_id'] + [c for c in target_cols if c in TAI_DB.columns]
print(TAI_DB[available_verify_cols].head(6))
```
